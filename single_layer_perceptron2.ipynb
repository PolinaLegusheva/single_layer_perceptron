```python
import numpy as np
import matplotlib.pyplot as plt
from perceptron import Perceptron
from plotting import Plotting
```

```python
polygon = np.array([[2, 2], [5, 8], [12, 6], [9, 1]])
points = np.array([[3, 4], [6, 6], [8, 3], [10,10]])

perceptron = Perceptron(input_size=2,  hidden_size=4, output_size=1)
perceptron.train(points, polygon)

plotting = Plotting()
plotting.plot_polygon(polygon)
plotting.plot_decision_boundary(perceptron.weights_output, perceptron.bias_output, min(points[:, 0]), max(points[:, 0]), 'Гиперплоскость')
point_in_polygon = np.array([point for point in points if perceptron.point_in_polygon(point, polygon)])
point_out_polygon = np.array([point for point in points if not perceptron.point_in_polygon(point, polygon)])
plotting.plot_points(point_in_polygon, point_out_polygon)
plotting.show_plot()
```
